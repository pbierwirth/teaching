# Kapitel 2: Listen
In Python kann man einer Variable auch mehrere Einträge zuordnen. Eine der gängigsten Formen, um mehrere Einträge in einer Variable zu hinterlegen/speichern ist die sogenannte **Liste**.  Listen kann man einfach über eckige Klammern erstellen:

```python
x = [1,2,3]
print(x)
```

Prinzipiell können Listen unterschiedliche Datentypen beinhalten (z.B. Strings, Floats, etc.):

```python
y = ['hallo', 1, 2.45, 'ich bin ein Listeneintrag']
print(f'Die fertige Liste: y = {y}')
```

Die Liste beinhaltet Einträge unterschiedlichen Typus, ist aber selbst als Variablentyp eine "**list**":

```python
list_type = type(y)
print(f'Unsere Liste zählt zum Variablentyp {list_type}')
```

Man kann einfach auf die einzelnen Listeneinträge zugreifen in dem man den gewünschten Eintrag indiziert:

```python
print(y[0])
```

Hier ist es zu beachten, dass Python einen sogenannten Nullindex hat. D.h. mit der 0 wählt man den ersten Eintrag aus, mit einer 1 den zweiten (y[1]) usw...

# Kapitel 3: For-Loops und Conditional-Statements

For-Loops (im Deutschen „Schleifen“) und Conditional-Statements („Bedingte Anweisungen“) sind die fundamentalsten Bausteine fast jeder Programmiersprache. Sie sind zentral für die Automatisierung repetitiver (sich wiederholender) Operationen und Aufgaben. 

```python
for i in y:
    print(i)
```

Die obige Schleife/Loop zeigt, wie der Index *i* bei jeder Iteration (bei jeder Wiederholung) einen neuen Eintrag aus unserer Liste **y** annimmt.

Conditional-Statements basieren auf sogenannten logischen **Bools** (True or False), die überprüfen, ob eine bestimmte Aussage wahr oder falsch ist und geben dies als *True* oder *False*:

```python
a = 1==1
b = 1==2
c = 2>3
print(a)
print(b)
print(c)
```

Diese Bools lassen sich auch mit logischen **AND** und **OR** Operatoren verknüpfen:

```python
a = 10 > 5 and 10 > 8
b = 10 > 5 and 10 > 12
c = 10 > 5 or 10 > 12
print(a)
print(b)
print(c)
```

 Eine klassische Aufgabe, bei der wir einen Loop in Kombination mit einem Conditional-Statement super verwenden könnten, ist die Auswahl aller Einträge in einer Liste (z.B. *y*), die beispielsweise einen String beinhalten: 

```python
for i in y:
    if type(i) == str: # <- conditional statement: if true print entry (entry has to be string to be printed)
        print(i)
```

Wenn wir nun eine neue Liste erstellen wollen, müssen wir zunächst die Liste initialisieren (d.h. wir erstellen eine neue leere Liste, in die wir alle Einträge speichern, die Strings sind).

```python
string_list = [] # <- Initalisierung der Liste

for i in y:
    if type(i) == str:
        string_list.append(i)

print(string_list)
```
