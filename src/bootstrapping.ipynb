## 1. Datengeneration

Die Schätzung eines Parameter aus einer Gesamtpopulation ist immer mit einer gewissen Unsicherheit verbunden, wenn wir nicht die komplette Gesamtpopulation erfassen. 

Angenommen wir haben eine Population mit $N=10.000.000$. Die durschnittliche Körpergröße beträgt $\mu=170$ und die Standardabweichung beträgt $\sigma=10$. Ziehen wir nun Stichproben mit einer bestimmten Stichprobengröße aus dieser Gesamtpopulation, wird der geschätzte Populationsparamter (z.B. die durchschnittliche Körpergröße) über die Stichproben hinweg varrieren. 

Programmiersprachen wie `Python` oder `R` können super dafür verwendet werden, um diese hypothetischen Szenarien zu simulieren und so ein tiefes, intuitives Verständnis für die zugrundeliegende Statistik zu bekommen. Zunächst wollen wir die oben beschriebene Gesamtpopulation generieren. Dafür verwenden wir Numpy:

```python
import numpy as np 

# Wir verwenden Numpys `random`-Modul und die darin enthaltene 
# `normal`-Funtkion, um 10.000.000 normalverteilte Beobachtungen
# mit einem Mittelwert von 170 und einer Standardabweichung 
# von 10 zu generieren
data = np.random.normal(loc = 170, scale = 10, size = 10000000)

# Wir lassen uns die ersten 10 Beobachtungen ausgeben
print(data[:10])

# Außerdem überprüfen wir den Mittelwert und die Standardabweichung
print('mu: ', np.mean(data))
print('sigma', np.std(data))
```

Die Datengeneration hat funktioniert. Wir können die Daten der Gesmatpopulation mithilfe des `Matplotlib` Packages visualisieren, um zu überprüfen, ob die Daten auch wirklich normalverteilt sind:

```python
import matplotlib.pyplot as plt

# Um einen schnellen Überblick zu erhalten verwenden wir die `hist`-Funktion.
# Als erstes Argument verwenden wir die generierten Daten. 
# Mit dem zweiten Argument `bins` legen wir fest, wie viele "Säulen"
# unser Histogram haben soll: 

plt.hist(data, bins = 100);
plt.show()
```

Bei einer Populationsgröße von 10.000.000 sieht die Verteilung der Daten normalverteilt aus. Wir haben also mit der Generation der Gesamtpopulation den Grundstein für unsere Stichproben-Simulation gelegt.

## 2. Sampling und Re-Sampling 

 Als nächstes wollen wir mehrere Stichproben aus der Gesamtpopulation ziehen, deren Mittelwerte bestimmen und abspeichern, damit wir sehen, wie die Stichprobenschätzungen variieren. Dafür können wir wieder auf Numpys `random` Modul zurückgreifen. Wir verwenden die `choice`-Funktion, um zufällige Beobachtungen aus der Gesamtpopualtion auszuwählen:

```python
# Zunächst definieren wir die Stichprobengröße n,
# die wir unten in der Choice-Funktion verwenden
n = 10

# Das erste Argument der Choice-Funktion beinhaltet die 
# Daten (als Liste oder Numpy-Array), aus denen einzelne Beobachtungen
# gezogen werden sollen. Das zweite Argument `size` sagt der Funktion
# wie viele Beobachtungen aus dem ersten Argument, also unseren Daten, 
# gezogen werden sollen
sample = np.random.choice(data,size=n) 

# wir lassen uns alle n ausgewählten Beobachtungen, deren Mittelwert
# und Standardabweichung ausgeben
print(sample)
print('mean: ', np.mean(sample))
print('std: ', np.std(sample))
```

Um ein erstes Gefühl für die Ungewissheit der Parameterschätzung zu ehalten, die mit einer bestimmten Stichpobegröße assoziiert ist, reicht es natürlich nicht nur eine Stichprobe aus der Gesamtpopulation zu ziehen. Wir müssen also den obigen Code wiederholen und die berechneten Stichprobenmittelwerte abspeichern. Dafür verwenden wir, wie im vorangegangen Kapitel, `for-Schleifen` und `Listen`. 

Im Prinzip können Sie die obige Zelle wiederholt abspielen und Sie werden bereits festellen, dass Mittelwert und Standardabweichung zwischen den Stichproben variieren. Wir wollen diesen Prozess aber nun automatisieren:


```python
# zunächst initialisieren wir eine leere Liste, in die wir die 
# berechneten Stichprobenmittelwerte ablegen können
mittelwerte = []

# Dann bestimmen wir die Anzahl der Stichproben, 
# die wir aus der Gesamtpopulation ziehen möchten
iterationen = 10000

# Und wir legen die Größe der Stichprobe fest
n = 10

# Wir bauen eine Schleife
for _ in range(iterationen):
   sample = np.random.choice(data, size=n)
   mittelwerte.append(np.mean(sample))
```

Wir haben in der obigen Schleife 10.000 Stichpoben (n=10) aus der Gesamtpopulation gezogen, den Mittelwert berechnet und in eine Liste gespeichert, die somit 10.000 Stichprobenmittelwerte beinhaltet. Wir können nun die Unsicherheit, die mit der Parameterschätzung einer n=10 großen Stichprobe verknüpft sind einfach über die Standardabweichung dieser Liste bestimmen:

```python
simulierter_sem = np.std(mittelwerte)
print(simulierter_sem)

# Außerdem können wir die Verteilung der Mittelwerte plotten
plt.hist(mittelwerte,100);
```

Die Standardabweichung der Stichproben-Mittelwerte, die wir oben berechnet haben, entspricht annähernd dem `Standardfehler des Mittels (SEM)`, bzw. es ist einfach eine simulationsbasierte oder numerische Form des `SEM`. Dies lässt sich leicht überprüfen, in dem man den `simulations-ermittelden SEM`  mit dem `analytischen SEM` vergleicht. Zur Erinnerung der `analytische SEM` wird wie folg berechnet:

$$SEM = \frac{\sigma_x}{\sqrt{n}}$$



```python
sem = 10/np.sqrt(n)
print(sem)
print(simulierter_sem)
```

Der simulierte/numerische und der analytische SEM sollten hier ziemlich ähnlich sein. Die oben beschriebene Simulation sollte das "mechanische" Verständnis, bzw. die Intuition für den SEM vertiefen. Um das Verständnis noch weiter zu vertiefen, bearbeiten Sie bitte die folgenden Aufgaben:

## 3. Bootstrapping

Das obige Beispiel hat illustriert, wie man die Ungewissheit der Parameterschätzung durch eine einfache Simulation quantifizieren kann. Allerdings ergibt sich im Forscheralltag ein großes Problem: Wir haben keinen Zugriff auf alle Werte der Gesamtpopulation (Und wenn wir es hätten, müssten wir auch keine Ungewissheit in der Parameterschätzung quantifizieren, da diese ohnehin nicht-existent wäre). Wir haben in der Regel eine Stichprobe mit n Beobachtungen, die im besten Fall möglichst repräsentativ die Gesamtstichprobe abbildet. Die Ungewissheit der Parameterschätzung (z.B. der Mittelwert) lässt sich hier natürlich analytisch über die obige Formel realisieren. Interssanterweise kann man die Ungewissheit aber auch numerisch, d.h. über eine Simulation quantifizieren. Da wir aus einer Stichprobe mit n Beobachtungen natürlich nicht ohne weiteres neue Stichproben mit der gleichen Stichprobengröße ziehen können, müssen wir uns hier anders helfen, nämlich über das sog `Bootstrapping`. Wir ziehen uns an unseren eigenen Schnürrsenkeln (siehe Baron von Münchausen) aus dem "Schlamassel", indem wir unsere "eigene Stichprobe" nehmen und unterschiedliche Stichproben mit n Beobachtungen bilden.

Dabei spielt das Konzept des Samplen mit `Replacement` eine zentrale Rolle:


```python
# Wir haben eine Stichprobe mit vier Werten
sample = [1,2,3,4]
```


Wir stellen uns vor, dass diese vier Werte zu vier Kugeln gehören, 
die in einer Box liegen. Samplen wir nun 4 Kugeln OHNE REPLACEMENT,
können wir jede Kugel einmal ziehen


```python
draws = np.random.choice(sample, size=4, replace=False)
print(draws)
```

Wenn wir mit `replace=False` öfter als vier-mal (`size > 4`) samplen würden, bekommen wir eine Fehlermeldung, da unsere "Box" leer ist.

```python
draws = np.random.choice(sample, size=5, replace=False)
print(draws)
```

Samplen wir nun allerdings **mit Replacement**, werfen wir die Kugeln quasi zurück in die Box nach jedem Draw und wir können im Prinzip unendlich viele Draws durchführen, was natürlich auch bedeutet, dass die Werte (oder Kugel) mehrfach vorkommen können:

```python
draws = np.random.choice(sample, size=5, replace=True)
print(draws)
```

Beim Bootstrapping macht man sich nun dieses `Sampling mit Replacement` zu nutze, um unterschiedliche Stichproben aus der Origninalstichprobe zu generieren. Der SEM lässt sich dann wie in unserer obigen Simulation über die Mittelwertsverteilung dieser simulierten Stichproben bestimmen:

```python
n = 100
data = np.random.normal(50,10,n) # Generation der Originalstichprobe
mittelwerte = []
iterationen = 10000

for _ in range(iterationen):
    # sample mit replacement aus Originalstichprobe
    sample =  np.random.choice(data, size=n, replace=True) 
    mittelwerte.append(np.mean(sample))

sem_bootstrapped = np.std(mittelwerte)
sem_analytical   = np.std(data)/np.sqrt(n)

print('bootstrapped SEM: ', sem_bootstrapped)
print('analytical SEM:   ', sem_analytical)
```

Bootstrapping hat den großen Vorteil, dass es nicht-parametrisch ist.  Zudem kann man Bootstrapping auch relativ leicht Einsetzen, um die Ungewissheit von Parameterschätzungen zu quantifzieren, für die es keine analytische Lösung gibt.

Das 95% Konfidenzintervall lässt sich ebenfalls über zwei Wege bestimmen:

```python
# analytisch

ci_an = [np.mean(data) - 1.96*sem_bootstrapped, 
    np.mean(data) + 1.96*sem_bootstrapped]

ci_bt = [np.percentile(mittelwerte, 2.5),
    np.percentile(mittelwerte, 97.5)]

print(ci_an)
print(ci_bt)
```

Das Konfidenzintervall in einer Studie sagt uns **NICHT**, dass der wahre Populationsparameter mit 95% Wahrscheinlichkeit in diesem Intervall liegt. Das Konfidenzintervall gibt uns Auskunft darüber, dass bei unendlich vielen Wiederholungen des gleichen Studiendesgins (gleiche Stichprobe, gleiche Varianz etc.) 95% der berechneten Konfidenzintervalle den wahren Wert beinhalten würden:

```python
data = np.random.normal(0,1,10000000)
n = 30

plt.figure()
for i in range(40):
    sample = np.random.choice(data, 30)
    mean = sample.mean()
    std = sample.std()
    ci = std/np.sqrt(n) * 1.96

    if (mean-ci > 0) or (mean+ci < 0):
        plt.scatter(mean, 41-i, color='red')
        plt.errorbar(mean, 41-i, xerr = ci, color='red')
    else:
        plt.scatter(mean, 41-i, color='black')
        plt.errorbar(mean, 41-i, xerr = ci, color='black')

plt.axvline(0, color='blue')
```
